```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
import string

# Set random seeds for reproducibility
tf.random.set_seed(42)
np.random.seed(42)

# Load the dataset
path = "https://github.com/robitussin/CCDEPLRL_EXERCISES/blob/9b8ac1c5683abecc144f0af47eb7cda0688e12b7/dataset/reviews.json?raw=true"
dataset = pd.read_json(path)
print("Dataset shape:", dataset.shape)
print("\nFirst 5 rows:")
print(dataset.head())

# ===== ENHANCED DATA PREPROCESSING =====
print("\n=== ENHANCED DATA PREPROCESSING ===")

def clean_text(text):
    """Enhanced text cleaning function"""
    if pd.isna(text):
        return ""

    # Convert to lowercase
    text = str(text).lower()

    # Remove HTML tags
    text = re.sub(r'<[^>]+>', '', text)

    # Remove URLs
    text = re.sub(r'http\S+', '', text)

    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text)

    # Remove punctuation but keep emoticons
    text = re.sub(r'[^\w\s:;(){}[\]<>!?.-]', '', text)

    return text.strip()

# Clean the reviews
reviews = [clean_text(review) for review in dataset['review'].tolist()]
ratings = dataset['rating'].tolist()

# Enhanced sentiment labeling - more balanced approach
labels = []
for rating in ratings:
    if rating >= 4:
        labels.append(1)  # Positive
    elif rating <= 2:
        labels.append(0)  # Negative
    else:
        # Skip neutral ratings for better classification
        continue

# Filter out neutral reviews
filtered_reviews = []
filtered_labels = []
for i, rating in enumerate(ratings):
    if rating != 3:  # Skip neutral ratings
        filtered_reviews.append(reviews[i])
        filtered_labels.append(1 if rating >= 4 else 0)

reviews = filtered_reviews
labels = filtered_labels

print(f"Filtered dataset size: {len(reviews)}")
print(f"Positive samples: {sum(labels)}")
print(f"Negative samples: {len(labels) - sum(labels)}")

# Enhanced train-validation-test split
X_temp, X_test, y_temp, y_test = train_test_split(
    reviews, labels, test_size=0.15, random_state=42, stratify=labels
)

X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.176, random_state=42, stratify=y_temp  # ~15% of total
)

print(f"\nTraining samples: {len(X_train)}")
print(f"Validation samples: {len(X_val)}")
print(f"Testing samples: {len(X_test)}")

# Convert to numpy arrays
y_train = np.array(y_train)
y_val = np.array(y_val)
y_test = np.array(y_test)

# ===== ENHANCED TOKENIZATION =====
print("\n=== ENHANCED TOKENIZATION ===")

# Increased vocabulary size and better parameters
vocab_size = 5000  # Increased from 1000
embedding_dim = 128  # Increased from 16
max_length = 150  # Increased from 100
trunc_type = 'post'
padding_type = 'post'
oov_tok = "<OOV>"

# Create and fit the tokenizer
tokenizer = Tokenizer(
    num_words=vocab_size,
    oov_token=oov_tok,
    filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n'  # Keep more characters
)
tokenizer.fit_on_texts(X_train)
word_index = tokenizer.word_index
print(f"Found {len(word_index)} unique tokens")

# Convert text to sequences
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_val_seq = tokenizer.texts_to_sequences(X_val)
X_test_seq = tokenizer.texts_to_sequences(X_test)

# Pad sequences
X_train_pad = pad_sequences(X_train_seq, maxlen=max_length,
                           padding=padding_type, truncating=trunc_type)
X_val_pad = pad_sequences(X_val_seq, maxlen=max_length,
                         padding=padding_type, truncating=trunc_type)
X_test_pad = pad_sequences(X_test_seq, maxlen=max_length,
                          padding=padding_type, truncating=trunc_type)

print(f"Training data shape: {X_train_pad.shape}")
print(f"Validation data shape: {X_val_pad.shape}")
print(f"Testing data shape: {X_test_pad.shape}")

# ===== COMPUTE CLASS WEIGHTS =====
class_weights = compute_class_weight(
    'balanced',
    classes=np.unique(y_train),
    y=y_train
)
class_weight_dict = dict(enumerate(class_weights))
print(f"Class weights: {class_weight_dict}")

# ===== ENHANCED MODEL ARCHITECTURE =====
print("\n=== BUILDING ENHANCED MODEL ===")

def create_model():
    model = tf.keras.Sequential([
        # Embedding layer with larger dimensions
        tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length=max_length),

        # Bidirectional LSTM layers
        tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(128,
                                                          dropout=0.3,
                                                          recurrent_dropout=0.3,
                                                          return_sequences=True)),
        tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64,
                                                          dropout=0.3,
                                                          recurrent_dropout=0.3)),

        # Dense layers with batch normalization
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.5),

        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.3),

        # Output layer
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])

    return model

model = create_model()

# Compile with optimized parameters
model.compile(
    loss='binary_crossentropy',
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    metrics=['accuracy', 'precision', 'recall']
)

print("\nEnhanced Model Architecture:")
model.summary()

# ===== ENHANCED TRAINING WITH CALLBACKS =====
print("\n=== TRAINING WITH CALLBACKS ===")

# Define callbacks
callbacks = [
    EarlyStopping(
        monitor='val_accuracy',
        patience=5,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=3,
        min_lr=1e-7,
        verbose=1
    ),
    ModelCheckpoint(
        'best_model.h5',
        monitor='val_accuracy',
        save_best_only=True,
        verbose=1
    )
]

# Train the model
print("\nTraining the enhanced model...")
num_epochs = 25  # Increased epochs with early stopping

history = model.fit(
    X_train_pad, y_train,
    epochs=num_epochs,
    batch_size=32,  # Reduced batch size for better convergence
    validation_data=(X_val_pad, y_val),
    class_weight=class_weight_dict,
    callbacks=callbacks,
    verbose=1
)

# ===== ENHANCED VISUALIZATION =====
print("\n=== ENHANCED TRAINING VISUALIZATION ===")

def plot_enhanced_graphs(history):
    plt.figure(figsize=(20, 12))

    # Accuracy plot
    plt.subplot(2, 3, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True, alpha=0.3)

    # Loss plot
    plt.subplot(2, 3, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True, alpha=0.3)

    # Precision plot
    plt.subplot(2, 3, 3)
    plt.plot(history.history['precision'], label='Training Precision')
    plt.plot(history.history['val_precision'], label='Validation Precision')
    plt.title('Model Precision')
    plt.xlabel('Epoch')
    plt.ylabel('Precision')
    plt.legend()
    plt.grid(True, alpha=0.3)

    # Recall plot
    plt.subplot(2, 3, 4)
    plt.plot(history.history['recall'], label='Training Recall')
    plt.plot(history.history['val_recall'], label='Validation Recall')
    plt.title('Model Recall')
    plt.xlabel('Epoch')
    plt.ylabel('Recall')
    plt.legend()
    plt.grid(True, alpha=0.3)

    # Learning rate plot (if available)
    plt.subplot(2, 3, 5)
    if 'lr' in history.history:
        plt.plot(history.history['lr'])
        plt.title('Learning Rate')
        plt.xlabel('Epoch')
        plt.ylabel('Learning Rate')
        plt.yscale('log')
        plt.grid(True, alpha=0.3)

    # F1 Score calculation and plot
    plt.subplot(2, 3, 6)
    precision = np.array(history.history['precision'])
    recall = np.array(history.history['recall'])
    f1_train = 2 * (precision * recall) / (precision + recall + 1e-7)

    val_precision = np.array(history.history['val_precision'])
    val_recall = np.array(history.history['val_recall'])
    f1_val = 2 * (val_precision * val_recall) / (val_precision + val_recall + 1e-7)

    plt.plot(f1_train, label='Training F1-Score')
    plt.plot(f1_val, label='Validation F1-Score')
    plt.title('F1-Score')
    plt.xlabel('Epoch')
    plt.ylabel('F1-Score')
    plt.legend()
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

plot_enhanced_graphs(history)

# ===== COMPREHENSIVE MODEL EVALUATION =====
print("\n=== COMPREHENSIVE MODEL EVALUATION ===")

# Evaluate on test set
test_results = model.evaluate(X_test_pad, y_test, verbose=0)
print(f"Test Loss: {test_results[0]:.4f}")
print(f"Test Accuracy: {test_results[1]:.4f}")
print(f"Test Precision: {test_results[2]:.4f}")
print(f"Test Recall: {test_results[3]:.4f}")

# Calculate F1-score
test_precision = test_results[2]
test_recall = test_results[3]
f1_score = 2 * (test_precision * test_recall) / (test_precision + test_recall)
print(f"Test F1-Score: {f1_score:.4f}")

# Detailed predictions analysis
test_predictions = model.predict(X_test_pad)
test_pred_binary = (test_predictions > 0.5).astype(int).flatten()

# Classification report
from sklearn.metrics import classification_report, confusion_matrix
print("\n=== DETAILED CLASSIFICATION REPORT ===")
print(classification_report(y_test, test_pred_binary,
                          target_names=['Negative', 'Positive']))

print("\n=== CONFUSION MATRIX ===")
cm = confusion_matrix(y_test, test_pred_binary)
print(cm)

# Save the model
model.save('enhanced_sentiment_model.h5')
print("\nModel saved as 'enhanced_sentiment_model.h5'")

# Generate embedding files for visualization
print("\n=== GENERATING ENHANCED VISUALIZATION FILES ===")
e = model.layers[0]
weights = e.get_weights()[0]
print(f"Embedding weights shape: {weights.shape}")

reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])

import io
out_v = io.open('enhanced_vecs.tsv', 'w', encoding='utf-8')
out_m = io.open('enhanced_meta.tsv', 'w', encoding='utf-8')

for word_num in range(1, min(vocab_size, len(word_index) + 1)):
    if word_num in reverse_word_index:
        word = reverse_word_index[word_num]
        embeddings = weights[word_num]
        out_m.write(word + "\n")
        out_v.write('\t'.join([str(x) for x in embeddings]) + "\n")

out_v.close()
out_m.close()

print("Generated enhanced_vecs.tsv and enhanced_meta.tsv")
print("Upload these files to http://projector.tensorflow.org/ to visualize word embeddings")

print(f"\n🎉 MODEL ENHANCEMENT COMPLETE!")
print(f"Target Accuracy: 90%")
print(f"Achieved Accuracy: {test_results[1]*100:.2f}%")
if test_results[1] >= 0.90:
    print("✅ TARGET ACCURACY ACHIEVED!")
else:
    print("⚠️  Close to target - consider additional fine-tuning")
```

```python
# ===== ENHANCED PREDICTION TESTING =====
print("\n=== ENHANCED PREDICTION TESTING ===")

# Enhanced Tagalog reviews for testing
new_tagalog_reviews = [
    "Hindi maganda ng produkto na ito!",  # Very beautiful product!
    "Hindi ko gusto ang pagkain dito",     # I don't like the food here
    "Masarap na pagkain at magandang serbisyo",  # Delicious food and good service
    "Pangit ang kalidad ng produkto",      # Bad quality product
    "Sulit na sulit ang pera ko dito",     # Very worth my money
    "Ayaw ko na bumalik sa lugar na ito",  # I don't want to return to this place
    "Maganda ang packaging at mabilis ang delivery",  # Nice packaging and fast delivery
    "Masama ang ugali ng mga staff",       # Bad attitude of the staff
    "Perfect fit at maganda ang tela",     # Perfect fit and nice fabric
    "Sobrang tagal ng delivery",           # Very long delivery time
    "Excellent service and quality!",      # English positive
    "Terrible experience, worst ever",     # English negative
    "Amazing product, highly recommended", # English positive
    "magaling si ericka"         # English negative
    ""
]

# Clean and process new reviews
cleaned_new_reviews = [clean_text(review) for review in new_tagalog_reviews]
new_sequences = tokenizer.texts_to_sequences(cleaned_new_reviews)
new_padded = pad_sequences(new_sequences, maxlen=max_length,
                          padding=padding_type, truncating=trunc_type)

# Make predictions
predictions = model.predict(new_padded)

print("\nEnhanced Sentiment Analysis Results:")
print("=" * 90)
for i, (original, cleaned) in enumerate(zip(new_tagalog_reviews, cleaned_new_reviews)):
    sentiment_score = predictions[i][0]
    sentiment = "POSITIVE" if sentiment_score > 0.5 else "NEGATIVE"
    confidence = sentiment_score if sentiment_score > 0.5 else 1 - sentiment_score

    print(f"Original: {original}")
    print(f"Cleaned:  {cleaned}")
    print(f"Sentiment: {sentiment} (Score: {sentiment_score:.4f}, Confidence: {confidence:.4f})")
    print("-" * 90)
```
